```python
from keras.layers import Input, Lambda, Dense, Flatten
from keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
```

```python
IMAGE_SIZE = [224, 224]

vgg = VGG16(
    input_shape = IMAGE_SIZE + [3], 
    weights = 'imagenet', 
    include_top = False
)

for layer in vgg.layers:
  layer.trainable = False

x = Flatten()(vgg.output)
prediction = Dense(102, activation = 'sigmoid')(x)
model = Model(inputs = vgg.input, outputs = prediction)

model.compile(
  loss='binary_crossentropy',
  optimizer='adam',
  metrics=['accuracy']
)

print(model.summary())
```

```python
train_path = '../input/ip02-dataset/classification/train'
valid_path = '../input/ip02-dataset/classification/val'
test_path = '../input/ip02-dataset/classification/test'

train_generator = ImageDataGenerator(rescale=1./255).flow_from_directory(directory=train_path, 
                                                                         class_mode='categorical', 
                                                                         batch_size=32,
                                                                         target_size=(224, 224),
                                                                         color_mode="rgb",
                                                                         shuffle=True)


# Load Test Data as image generator with Keras, Normalization with (rescale=1./255)
val_generator = ImageDataGenerator(rescale=1./255).flow_from_directory(directory=valid_path, 
                                                                        class_mode='categorical', 
                                                                        batch_size=32,
                                                                        target_size=(224, 224),
                                                                        color_mode="rgb",
                                                                        shuffle=False)

test_generator = ImageDataGenerator(rescale=1./255).flow_from_directory(directory=test_path, 
                                                                        class_mode='categorical', 
                                                                        batch_size=32,
                                                                        target_size=(224, 224),
                                                                        color_mode="rgb",
                                                                        shuffle=False)
```

```python
r = model.fit_generator(
  train_generator,  
  validation_data = val_generator,
  epochs = 20,
  steps_per_epoch = (len(train_generator) // 32),
  validation_steps = (len(val_generator) // 32)
)
```

```python
loss, accuracy = model.evaluate(test_generator, verbose=1)
# loss_v, accuracy_v = model.evaluate(x_validate, y_validate, verbose=1)
# print("Validation: accuracy = %f  ;  loss_v = %f" % (accuracy_v, loss_v))
print("Test: accuracy = %f  ;  loss = %f" % (accuracy, loss))
```
