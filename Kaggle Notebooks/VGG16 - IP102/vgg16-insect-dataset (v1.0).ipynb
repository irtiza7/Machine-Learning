```python
import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
```

```python
image_rows = 224
image_cols = 224
image_dim = 3
image_shape  = (image_rows, image_cols, image_dim)
```

```python
# Creating VGG16 Model from scratch
model = Sequential()

## Block 1 with 2 Conv Layers and 64 filters
conv1_1 = Conv2D(
    input_shape = image_shape,
    filters = 64,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv1_2  = Conv2D(
    filters = 64,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
pool_1 = MaxPool2D(
    pool_size = (2, 2),
    strides = (2, 2)
)

model.add(conv1_1)
model.add(conv1_2)
model.add(pool_1)

## Block 2 with 2 Conv Layers and 128 filters
conv2_1 = Conv2D(
    filters = 128,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv2_2 = Conv2D(
    filters = 128,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
pool_2 = MaxPool2D(
    pool_size = (2, 2),
    strides = (2, 2)
)

model.add(conv2_1)
model.add(conv2_2)
model.add(pool_2)

## Block 3 with 3 Conv Layers and 256 filters
conv3_1 = Conv2D(
    filters = 256,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv3_2 = Conv2D(
    filters = 256,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv3_3 = Conv2D(
    filters = 256,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
pool_3 =  MaxPool2D(
    pool_size = (2, 2),
    strides = (2, 2)
)

model.add(conv3_1)
model.add(conv3_2)
model.add(conv3_3)
model.add(pool_3)

## Block 4 with 3 Conv Layers and 512 filters
conv4_1 = Conv2D(
    filters = 512,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv4_2 = Conv2D(
    filters = 512,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv4_3 = Conv2D(
    filters = 512,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
pool_4 =  MaxPool2D(
    pool_size = (2, 2),
    strides = (2, 2)
)

model.add(conv4_1)
model.add(conv4_2)
model.add(conv4_3)
model.add(pool_4)

## Block 5 with 3 Conv Layers and 512 filters
conv5_1 = Conv2D(
    filters = 512,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv5_2 = Conv2D(
    filters = 512,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
conv5_3 = Conv2D(
    filters = 512,
    kernel_size = (3, 3),
    padding = 'same',
    activation = 'relu'
)
pool_5 =  MaxPool2D(
    pool_size = (2, 2),
    strides = (2, 2)
)

model.add(conv5_1)
model.add(conv5_2)
model.add(conv5_3)
model.add(pool_5)

## Block 6 with 2 Dense Layers and Output Layer
dense_1 = Dense(
    units = 4096,
    activation = 'relu'
)
dense_2 = Dense(
    units = 4096,
    activation = 'relu'
)
output_layer = Dense(
    units = 1,
    activation = 'sigmoid'
)

model.add(Flatten())
model.add(dense_1)
model.add(dense_2)
model.add(output_layer)
```

```python
print(model.summary())
```
